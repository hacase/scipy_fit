# plot

```python
x=np.arange(0,12,1)
y=np.array([1, 1, 3, 2, 4, 5, 6, 8, 9, 10, 10, 11])
xerr=np.sqrt(x)
yerr=np.sqrt(y)

def lin(x, a, b):
    return x * a + b

fit.saveplot(fig_width=5.5)
fig, ax = plt.subplots()
ax.errorbar(x,y,yerr=yerr,ms=3,mew=0.5,marker="x",ls='',lw=0.5,capsize=2,label='Werte')
ax.plot(x,lin(x,0.961,0.4),lw=0.5,label='Fit')
ax.set_xlabel('Monat')
ax.set_ylabel('Temperatur [$\si{\degree}$C]')
plt.savefig("pic.png")
plt.show()
fit.revert_params()
```

```python
fit.exefit(lin,x,y,yerr=yerr,copy=1,flag=1)
```

```python
fit.exefit(lin,x,y,xerr=xerr,yerr=yerr,copy=1,flag=1)
```
